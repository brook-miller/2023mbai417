```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

from datetime import datetime, timedelta
from dateutil.parser import parse
```

```python
subscribers = pd.read_csv('https://raw.githubusercontent.com/brook-miller/msai339-students/main/tidy/subscribers.csv')
subscribers['subscribe_date'] = pd.to_datetime(subscribers['date'])
```

```python
months = pd.DataFrame()
months["month"] = pd.date_range('2022-01-01','2022-12-01', freq='MS').tolist()
```

```python
joined = subscribers.join(months, how="cross")
joined = joined.loc[joined["month"] >= joined["date"]]
joined = joined.reset_index(drop = True)
joined
```

```python
monthly_revenue = joined.groupby(by="month")['monthly_amount'].sum().reset_index()
monthly_revenue
```
